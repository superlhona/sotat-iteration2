```python
customers_orders = pd.read_csv("telecom_users.csv")
```

```python
customers_orders.drop(customers_orders.loc[customers_orders["Churn"]=="Yes"].index, inplace=True)
customers_orders.head()
```

```python
customers_orders.info()
```

```python
users = customers_orders[["customerID"]]
users["photoID"] = 0
users["photoID"][0::4] = 1
users["photoID"][1::4] = 2
users["photoID"][2::4] = 3
users["photoID"][3::4] = 4
users.head()
```

```python
id = users['photoID'].tolist()
with open('users.json', 'w') as f:
    json.dump(id, f)
```
